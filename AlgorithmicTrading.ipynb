## Strategy

Combine Momentum and volatility metrics to pick stocks on a daily basis from 29 Dow components. The Strategy invests in 10 stocks per day, with all stocks having equal weight in the portfolio. On a daily basis each stock receives a score that combines its return ranking and volatility ranking. The return ranking system ranks stocks based on their returns for the last six month period. The volatily ranking system ranks stocks based on their volatility for the last six month period. Volatility is derived from taking the standard deviation of returns for the last six months. Stocks with lower volatility have a higher ranking, stocks with higher returns have a higher ranking. The rankings for each category are combined to have a final ranking, and the top ten ranking stocks make up the portfolio in equal weights. The stocks are chosen with the end of day price, and the chosen stocks are invested in the market the next day.

```python
# staritng capital
capital = 1
# equity_curve records capital over time
equity_curve = []

# loop to test strategy
for i in range(125, len(Return_Data_Frame) - 1):
    components = []
    # Dictionary contains the returns for the last six months sorted from higest to lowest
    Price_Rank = {}
    # Dictionary contains the volatilities for the last six months sorted from lowest to highest
    Vol_Rank = {}
    # Dictionary for combination of returns and volatility
    Price_Vol = {}
    # get return and volatility information for each stock daily
    for ticker in DowTickers:
        Price_Rank[ticker] = Price_Data_Frame[ticker][i] / Price_Data_Frame[ticker][i - 125]
        Vol_Rank[ticker] = np.std(Return_Data_Frame[ticker][i - 125: i ])
    # Sort both dictionaries
    Price_Rank = OrderedDict(sorted(Price_Rank.items(), key = itemgetter(1), reverse = True))
    Vol_Rank = OrderedDict(sorted(Vol_Rank.items(), key = itemgetter(1), reverse = False))
    j = 1
    # Create the momentum and volatility score
    for key in Price_Rank:
        Price_Vol[key] = j
        j += 1
    j = 1
    for key in Vol_Rank:
        Price_Vol[key] += j
        j += 1
    # Sort the scores
    Price_Vol = OrderedDict(sorted(Price_Vol.items(), key = itemgetter(1), reverse = False))
    j = 1
    for key in Price_Vol:
        ticker = str(key)
        # break out of the loop once the top ten stocks are selected
        if j == 11:
            break
        else:
            components.append((capital * 0.10) * Return_Data_Frame[ticker][i + 1])
            j += 1
    capital = sum(components)
    equity_curve.append(capital)
    
```
